# **Data Cleaning**

The data that we are reading in will be used to understand basketball trends and changes in game play over the years. We are mainly observing the three point attempt, '3P', column to see these changes per season. The first two rows to display and run the dataset have been removed as they were posing issues. The first row was blank and the other was just a label; those were being counted as cells in the csv file and was showing up in the graphs. They are not needed in our dataset. 

Below, we want to fill in the missing values with the mean values since the values range similarly. Write the code in order to this using the fillna() and mean() functions.

```python
NBAdata_df.fillna(NBAdata_df.mean())
```

## More Data Cleaning! ðŸ§¹
Over here we want to delete row 1 for year 2022 because the year is still going and the data is incomplete for the current year. Write the code in order to this using the iloc[] method.

```python
NBAdata_df = NBAdata_df.iloc[1: , :]
print("Modified Dataframe : ")
print(NBAdata_df)
```

**ðŸ“Š EDA (Exploratory Data Analysis) ðŸ“Š**

Here we want to analyze the trend of 3 point attempts over the years. We can do this using a bar graph. First, we want to label the x axis as the NBA season and the y axis as 3 Point Attempts. Lastly we want to title the bar graph. We can do all of this using the plt function from the matplotlib library. Write the code below to create the bar graph we want.

```python
from matplotlib.pyplot import figure

#Horizontal Bar graph (year vs 3 point attempts)
plt.bar(NBAdata_df['Season'], NBAdata_df['3PA'])
plt.xlabel('NBA Season')
plt.ylabel('3 Point Attempts')
plt.title('3 Point Attempts over the years')
plt.show()
```

The next example is more challenging but very crucial in data science. It is known as linear regression and is used in making predications based off the data that is already provided. This also helps when data is missing and can help create estimates of what it might have been. This is just shown for you guys to see an example of what this may look like. Try to understand and analyze what you think this visualization is trying to represent.

```python
#Using linear regression 3P in the future seasons
from sklearn.linear_model import LinearRegression
tpa_train = NBAdata_df
tpa_test = NBAdata_df
tpa_train.plot.scatter(x="Season", y="3P",fontsize = 20,figsize = (25,10))

tpa_train = tpa_train[tpa_train['3P'].notna()]
tpa_test = tpa_train[tpa_train['3P'].notna()]
tpa_X_train = tpa_train[["Season"]]
tpa_X_test = tpa_test[["Season"]]
tpa_y_train = tpa_train["3P"]
lin_model = LinearRegression()
lin_model.fit(X=tpa_X_train, y=tpa_y_train)
lin_model.predict(X=tpa_X_test)
```
